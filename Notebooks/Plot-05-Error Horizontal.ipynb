```python
gswp3_month_temp_df = gswp3_daily_temp_df.resample(rule="M").sum() 
gswp3_month_temp_df
```

```python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt

# Define las configuraciones
settings = {
    'binwidth': 5,
    'lims': [-100, 100]  # Rango para los gráficos
}

# Convierte el DataFrame en un arreglo de NumPy
data_array = gswp3_month_temp_df.to_numpy()

# Extrae las fechas y los nombres de modelos
dates = data_array[:, 0]
model_names = gswp3_month_temp_df.columns[:-1]  # Excluir la columna "Obs"

# Inicializa un DataFrame para almacenar los errores
error_df = pd.DataFrame(data_array[:, -11], columns=['Obs'])  # Columna "Obs"
for i, model in enumerate(model_names):
    error_df[model] = data_array[:, i + 1] - data_array[:, -11]

# Realiza la regresión lineal para cada modelo
regressions = {}
for model in model_names:
    slope, intercept, rvalue, pvalue, std_err = stats.linregress(error_df['Obs'], error_df[model])
    regressions[model] = {
        'slope': slope,
        'intercept': intercept,
        'rvalue': rvalue,
        'pvalue': pvalue,
        'std_err': std_err
    }

# Crea gráficos y realiza análisis de estadísticas para cada modelo
for model in model_names:
    chain_error = error_df[model]
    slope = regressions[model]['slope']
    intercept = regressions[model]['intercept']
    rvalue = regressions[model]['rvalue']
    pvalue = regressions[model]['pvalue']
    std_err = regressions[model]['std_err']
    
    R2 = rvalue**2
    rmse = np.sqrt(np.mean((chain_error)**2))
    mean = np.mean(chain_error)
    std = np.std(chain_error)
    q90 = np.percentile(np.abs(chain_error), 90)
    
    fig, ax = plt.subplots(1, 2, figsize=(15, 5), tight_layout=True)
    # Histograma
    ax[0].grid(which='major', linestyle=':', color='0.5')
    ax[0].axvline(x=0, ls='--', lw=1.5, color='k')
    binwidth = settings['binwidth']
    bins = np.arange(min(chain_error), max(chain_error) + binwidth, binwidth)
    density = ax[0].hist(chain_error, bins=bins, density=True, color='0.6', edgecolor='k', alpha=0.5)
    mu, std = stats.norm.fit(chain_error)
    pval = stats.normaltest(chain_error)[1]
    xlims = settings['lims']
    x = np.linspace(xlims[0], xlims[1], 100)
    p = stats.norm.pdf(x, mu, std)
    ax[0].plot(x, p, 'r-', linewidth=1)
    ax[0].set(xlabel='error [m]', ylabel='pdf', xlim=settings['lims'], title=model)
    str_stats = ' rmse = %.1f\n mean = %.1f\n std = %.1f\n q90 = %.1f' % (rmse, mean, std, q90)
    ax[0].text(0, 0.98, str_stats, va='top', transform=ax[0].transAxes, fontsize=14)
    # Boxplot
    data = [chain_error]
    ax[1].yaxis.grid()
    bp = ax[1].boxplot(data, 0, 'k.', labels=[model], patch_artist=True)
    for median in bp['medians']:
        median.set(color='k', linewidth=1.5)
    ax[1].set(ylabel='error [m]', ylim=settings['lims'])

# Muestra los gráficos
plt.show()
```
